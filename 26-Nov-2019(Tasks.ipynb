#### Read the remaing four csv files.Store them in cust_df,prod_df,ship_df,order_df

```python
cust_df =pd.read_csv('https://raw.githubusercontent.com/sumathi16/MSTP-MTP-L3/master/Datasets/global_sales_data/cust_dimen.csv')
prod_df =pd.read_csv('https://raw.githubusercontent.com/sumathi16/MSTP-MTP-L3/master/Datasets/global_sales_data/prod_dimen.csv')
ship_df =pd.read_csv('https://raw.githubusercontent.com/sumathi16/MSTP-MTP-L3/master/Datasets/global_sales_data/shipping_dimen.csv')
order_df=pd.read_csv('https://raw.githubusercontent.com/sumathi16/MSTP-MTP-L3/master/Datasets/global_sales_data/orders_dimen.csv')
```

###  Merge the market_fact and cust_df files and store the result  in df1

```python
df1 = pd.merge(mr_fact,cust_df)
df1
```

### Merge the df1 and prod_df files and store the result  in df2

```python
df2=pd.merge(df1,prod_df)
df2
```

###  Merge the df2 and ship_df files and store the result  in df3

```python
df3 = pd.merge(df2,ship_df)
df3
```

###  Merge the df3 and order_df files and store the result  in  master_df

```python
master_df = pd.merge(df3,order_df)
```
